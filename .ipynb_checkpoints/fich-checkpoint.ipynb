```python
# formating dates
from dateutil import parser

def standardize_dates(date_series, target_format="%Y-%m-%d"):
    def parse_date(date):
        try:
            return parser.parse(date).strftime(target_format)
        except Exception as e:
            return None 

    return date_series.apply(parse_date)

# Standardize the dates
df['strartingOrAsgnmtDate'] = standardize_dates(df['strartingOrAsgnmtDate'])

print(df['strartingOrAsgnmtDate'])
```

```python
# formating Times
from datetime import datetime

# Standardize time format (24-hour)
def standardize_time(time_str):
    try:
        # Parse time, infer 12/24-hour format
        parsed_time = datetime.strptime(time_str, "%I:%M:%S %p") if "AM" in time_str or "PM" in time_str else datetime.strptime(time_str, "%H:%M:%S")
        return parsed_time.strftime("%H:%M:%S")
    except ValueError:
        return None 

df['Hour'] = df['Hour'].apply(standardize_time)

print(df['Hour'])
```

# Data Preprocessing

```python
# import libs
import pandas as pd
```

```python
# import our dataset
df = pd.read_csv("MLOps_Acadimic__DataSET.csv")
df.head()
```

```python
# checking the shape of the data 
df.shape
```

```python
# removing the empty columns
df.drop(['Finish Time','Email'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# removing the rows where there is no "Professors" and fill the empty elements
df.dropna(subset=['Professors'],inplace=True)
```

```python
df.fillna({'GainedSkill': 'No Skill', 'Email': 'Unknoun','RequiredSkill': 'Unknoun','ProjectTitle':'Unknoun'}, inplace=True)
```

```python
df.isnull().sum()
```

```python
# formating the dates
from dateutil import parser

def standardize_dates(date_series, target_format="%Y-%m-%d"):
    def parse_date(date):
        try:
            return parser.parse(date).strftime(target_format)
        except Exception as e:
            return None 

    return date_series.apply(parse_date)

# standardize the dates
df['strartingOrAsgnmtDate'] = standardize_dates(df['strartingOrAsgnmtDate'])

print(df['strartingOrAsgnmtDate'])
```

```python
df.isnull().sum()
```

```python
# formating times
from datetime import datetime

# Standardize time format (24-hour)
def standardize_time(time_str):
    try:
        # Parse time, infer 12/24-hour format
        parsed_time = datetime.strptime(time_str, "%I:%M:%S %p") if "AM" in time_str or "PM" in time_str else datetime.strptime(time_str, "%H:%M:%S")
        return parsed_time.strftime("%H:%M:%S") 
    except ValueError:
        return None 

df['Hour'] = df['Hour'].apply(standardize_time)

print(df['Hour'])
```

```python
#Preprocess
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

# Initialize tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Define preprocessing function
def preprocess_text(text):
    if pd.isnull(text): 
        return ""
    text = re.sub(r'[^a-zA-Z\s]', '', text).lower()  # Remove special characters and digits
    words = [lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words]
    return ' '.join(words)

# Preprocess textual features
df['RequiredSkill'] = df['RequiredSkill'].apply(preprocess_text)
df['SchoolDepartment'] = df['SchoolDepartment'].apply(preprocess_text)
df['ProjectTitle'] = df['ProjectTitle'].apply(preprocess_text)
df['GainedSkill'] = df['GainedSkill'].apply(preprocess_text)
# Combine textual features
df['Combined_Text'] = df['RequiredSkill'] + ' ' + df['SchoolDepartment'] + ' ' + df['ProjectTitle']

```

```python
def clean_data(df):
    # Preprocess textual features
    for col in ['RequiredSkill', 'SchoolDepartment', 'ProjectTitle', 'GainedSkill']:
        if col in df.columns:
            df[col] = df[col].apply(preprocess_text)
        else:
            df[col] = ""  # Add column if missing
    
    # Combine textual features
    df['Combined_Text'] = (
        df['RequiredSkill'] + ' ' +
        df['SchoolDepartment'] + ' ' +
        df['ProjectTitle']
    )
    
    return df
```

```python
# Clean your data
df = clean_data(df)

# Save the cleaned data to a CSV file
output_file = "cleaned_data.csv"
df.to_csv(output_file, index=False)

print(f"Cleaned data has been saved to {output_file}.")
```

```python
# TF-IDF with n-grams (1-gram and 2-grams)
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=1000, ngram_range=(1, 2))
text_features = tfidf_vectorizer.fit_transform(df['Combined_Text'])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Example: Encoding professor expertise
df['ProfessorEncoded'] = label_encoder.fit_transform(df['Professors'])

# Concatenate numeric features with text embeddings
text_features_dense = text_features.toarray()
numeric_features = df[['ProfessorEncoded']].values  # Include other numeric features as necessary
import numpy as np
combined_features = np.hstack((text_features_dense, numeric_features))
```

```python
from sklearn.decomposition import TruncatedSVD

# Try different numbers of components
n_components = 50  # Adjust based on your experimentation
svd = TruncatedSVD(n_components=n_components, random_state=42)
reduced_features = svd.fit_transform(combined_features)
```

```python
from sklearn.mixture import GaussianMixture

# Fit GMM
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(reduced_features)

# Assign GMM labels
df['Cluster'] = gmm_labels
```

```python
from sklearn.cluster import AgglomerativeClustering

# Fit Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_labels = agg_clustering.fit_predict(reduced_features)

# Assign Agglomerative Clustering labels
df['Cluster'] = agg_labels
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Test multiple values for n_clusters
best_kmeans_score = -1
best_kmeans_model = None

for n_clusters in range(2, 10):  # Try between 2 and 5 clusters
    kmeans = KMeans(n_clusters=n_clusters, n_init=20, random_state=42)
    labels = kmeans.fit_predict(reduced_features)
    
    # Evaluate metrics
    silhouette = silhouette_score(reduced_features, labels)
    calinski = calinski_harabasz_score(reduced_features, labels)
    davies = davies_bouldin_score(reduced_features, labels)

    print(f"Clusters: {n_clusters}, Silhouette: {silhouette}, Calinski-Harabasz: {calinski}, Davies-Bouldin: {davies}")
    
    if silhouette > best_kmeans_score:
        best_kmeans_score = silhouette
        best_kmeans_model = kmeans

# Use the best model
df['Cluster'] = best_kmeans_model.labels_
```

```python
from sklearn.cluster import KMeans

# Fit KMeans with 4 clusters
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
kmeans.fit(reduced_features)

# Assign cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Evaluate the final clustering performance
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

silhouette = silhouette_score(reduced_features, df['Cluster'])
calinski = calinski_harabasz_score(reduced_features, df['Cluster'])
davies = davies_bouldin_score(reduced_features, df['Cluster'])

print(f"Optimal Clusters: {optimal_clusters}")
print(f"Silhouette Score: {silhouette}")
print(f"Calinski-Harabasz Score: {calinski}")
print(f"Davies-Bouldin Score: {davies}")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(reduced_features)

# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.title("Visualization of Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# Fonction pour rechercher les recommandations
def recommender(student_skills, student_department, start_date, num_recommendations=5):
    """
    Recommande les cours les plus pertinents Ã  un Ã©tudiant en fonction de ses compÃ©tences et de son dÃ©partement.
    
    ParamÃ¨tres :
    - student_skills (list): CompÃ©tences de l'Ã©tudiant.
    - student_department (str): DÃ©partement acadÃ©mique de l'Ã©tudiant.
    - start_date (str): Date de dÃ©but souhaitÃ©e ('YYYY-MM-DD').
    - num_recommendations (int): Nombre de recommandations Ã  retourner.
    
    Retour :
    - DataFrame des cours recommandÃ©s.
    """
    # Filtrer les cours dans le mÃªme cluster
    filtered_df = df.copy()
    
    # Filtrer par dÃ©partement
    if student_department:
        filtered_df = filtered_df[filtered_df['SchoolDepartment'] == student_department]
    
    # VÃ©rifier les compÃ©tences de l'Ã©tudiant
    if student_skills:
        filtered_df['SkillMatch'] = filtered_df['RequiredSkill'].apply(
            lambda x: sum(skill in x for skill in student_skills)
        )
        filtered_df = filtered_df[filtered_df['SkillMatch'] > 0]
    
    # Filtrer par date de dÃ©but
    if start_date:
        filtered_df = filtered_df[filtered_df['strartingOrAsgnmtDate'] >= start_date]
    
    # Trier les rÃ©sultats par pertinence (SkillMatch puis cluster)
    filtered_df = filtered_df.sort_values(by=['SkillMatch', 'Cluster'], ascending=[False, True])
    
    # Retourner les meilleurs rÃ©sultats
    return filtered_df.head(num_recommendations)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load cleaned data
file_path = "cleaned_data.csv"  # Update with the actual path to your file
df = pd.read_csv(file_path)

# Vectorize the textual data
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['Combined_Text'])

# Recommendation function
def recommend_courses(input_text, top_n=5):
    """
    Recommends courses based on the input text.
    
    Parameters:
    - input_text (str): The user input describing skills, department, etc.
    - top_n (int): Number of top recommendations to return.
    
    Returns:
    - DataFrame: Top recommended courses with details.
    """
    input_vector = vectorizer.transform([input_text])

    similarity_scores = cosine_similarity(input_vector, tfidf_matrix)

    top_indices = similarity_scores[0].argsort()[-top_n:][::-1]
    recommendations = df.iloc[top_indices]
    
    return recommendations[['Professors', 'ProjectTitle', 'RequiredSkill']]

# Test the recommendation system
user_input = "machine learning computer science"  
recommended_courses = recommend_courses(user_input, top_n=5)


print("Recommended Courses:")
print(recommended_courses)
```
